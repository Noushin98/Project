**Data Visualization 3**

```python
Code_Value = ['DT.UND.DPPG.CD']
df3 = df[df['Series_Code'].isin(Code_Value)]
```

```python
# Group the data by region/continent to get the total undisbursed external debt for each region/continent
df3_region = df3.groupby('Country_Name', as_index=False)[2021].sum()

# Create a geographical graph to visually compare the total undisbursed external debt for each region/continent
fig = px.choropleth(df3_region, 
                    locations='Country_Name', 
                    locationmode='country names',
                    color=2021,
                    color_continuous_scale='Blues', 
                    range_color=(0, 60000000000),
                    title='Total undisbursed external debt by region/continent')
fig.show()
```

```python
# Use pie charts to analyze which countries' income groups contribute to the region's overall performance
df3_income = df3.groupby(['Region', 'Income_Group'], as_index=False).sum(numeric_only=True)
for region in df3_income['Region'].unique():
    df3_income_region = df3_income[df3_income['Region']==region]
    fig = px.pie(df3_income_region, values=2021, names='Income_Group', title=f'{region} - Undisbursed External Debt Total by Income Group')

fig.show()
```

**Data Visualization 6**

```python
Code_Value = ['FI.RES.TOTL.CD']
df6 = df[df['Series_Code'].isin(Code_Value)]
```

```python
df6 = df6.sort_values([2021], ascending=[False])
```

```python
df6 = df6.drop(df6[df6['Country_Code'].isin(['EAP','ECA', 'IDX', 'IDA', 'LAC', 'LMY', 'LIC','LMC','MNA', 'MIC','SAS','UMC', 'SSA', 'LDC'])].index)
```

```python
data5 = df6.iloc[0:5, 12:23]
```

```python
data5 = data5.assign(Country_Name= df6['Country_Name'])
```

```python
data5
```

```python
# replace 'Russian Federation' with 'Russian' to avoid country_name overlaps
data5['Country_Name'] = data5['Country_Name'].replace({'Russian Federation': 'Russia'})

# melt the dataframe to create long format data
df_melt = pd.melt(data5, id_vars=['Country_Name'], var_name='Year', value_name='Total Reserve')

# set a custom color palette
colors = sns.color_palette('Set2', n_colors=len(data5['Country_Name'].unique()))

# create the facet bar plot
g = sns.FacetGrid(df_melt, col='Year', col_wrap=3, height=4, aspect=1)
g.map_dataframe(sns.barplot, x='Country_Name', y='Total Reserve', palette=colors)
g.set_titles(col_template='Year {col_name}')
g.set_axis_labels('Country_Name', 'Total Reserve')

g.tight_layout()
sns.despine()

# show the plot
g.fig.suptitle('Total Reserve by Country and Year', y=1.05, fontsize=16)
plt.show()
```
